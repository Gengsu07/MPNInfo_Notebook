```python
import pandas as pd
```

```python
df=pd.read_csv(r'D:\DATA KANTOR\MPN INFO\afterSMO\2020.csv', sep='|',
dtype={'NPWP_FULL':'str','kdmap':'str','kdbayar':'str','masa':'str','tahun':'str'},parse_dates = ['datebayar'])
```

```python
mf = pd.read_excel('D:\DATA KANTOR\MASTERFILE\MFWP 2020 Kompilasi.xlsx', sheet_name='Nov-Des',usecols=['NPWP FULL','Kode KLU','Status WP'], 
dtype={'NPWP FULL':'str', 'Kode KLU':'str'})
```

```python
df = pd.merge(df,mf, left_on='NPWP_FULL', right_on='NPWP FULL', how='left')
```

```python
df.drop(['nosk','nospm','ket','NPWP FULL'],axis=1, inplace=True)
```

```python
df.head()
```

```python
map = pd.read_excel('D:\DATA KANTOR\MASTERFILE\KODE_MAP_FILTERED.xlsx', engine='openpyxl', dtype={'KD MAP':'str'})
```

```python
df = pd.merge(df,map, left_on='kdmap', right_on='KD MAP',how='left')
```

```python
klu = pd.read_csv('D:/DATA KANTOR/MASTERFILE/KLU.csv', sep=';',usecols=['klu_kode','klu_sektor'])
klu_kategori = pd.read_excel('D:/DATA KANTOR/MASTERFILE/KLU_sektor.xlsx', engine='openpyxl' )
```

```python
df = pd.merge(df,klu, left_on='Kode KLU',right_on='klu_kode')
df = pd.merge(df, klu_kategori, left_on='klu_sektor',right_on='Kode')
```

```python
df.head()
```

```python
df.drop(['klu_kode','klu_sektor'],axis=1,inplace=True)
```

```python
df = df[df['datebayar']< '2020-10-01']
```

```python
df.to_excel(r'D:\DATA KANTOR\MPN INFO\afterSMO\2020_jan_sept.xlsx',index=False)
```
