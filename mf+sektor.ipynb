```python
import pandas as pd
```

```python
mf = pd.read_excel(r'E:\NEXTCLOUD\DATA\DATABASE\Masterfile-007-2022.xlsx',sheet_name='FAST', 
dtype={'FULL':'str','KODE_KLU':'str'})
```

```python
klu = pd.read_csv(r'E:\NEXTCLOUD\DATA\DATABASE\KLU.csv', dtype={'klu_sektor':'str','klu_kode':'str'},sep=';')
```

```python
mf_klu = pd.merge(mf,klu,left_on='KODE_KLU',right_on='klu_kode',how='left')
```

```python
sektor = pd.read_excel(r'E:\NEXTCLOUD\DATA\DATABASE\KLU_sektor.xlsx')
```

```python
mf_sektor = pd.merge(mf_klu,sektor, left_on='klu_sektor',right_on='Kode',how='left')
```

```python
mf_sektor.drop(['klu_kode', 'klu_nama', 'klu_sektor'],inplace=True,axis=1)
```
